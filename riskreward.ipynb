```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Dow30.csv")
data.head()
```

```python
ticker = data["Symbol"].to_list()
ticker
```

```python
stocks = yf.download(ticker, start = "2010-01-01",end = "2020-01-01")
stocks
```

```python
stocks.describe()
```

```python
close = stocks.loc[:,"Close"]
close
```

```python
close = stocks.loc[:,"Close"]
close.plot(figsize=(15,8),fontsize = 12)
plt.legend(fontsize=12)
plt.show()
```

```python
normclose = close.div(close.iloc[0]).mul(100)
```

```python
ret = close.pct_change().dropna().mul(100)
ret
```

```python
summary = ret.describe().T.loc[:,["mean","std"]]
summary
```
